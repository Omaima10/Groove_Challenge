# Now im gonna extract the audio features and dump these features into csv file.

***The main block chain to extract feature from an audio is divided in to three parts as pre-emphasis, frame blocking & windowing and feature extraction to come out with feature coefficient.
After reading some papers about Audio features ectraction such as (https://www.ijitee.org/wp content/uploads/papers/v8i7c2/G10460587C219.pdf https://www.researchgate.net/publication/332254069_Robust_Feature_Extraction_Techniques_in_Speech_Recognition_A_Comparative_Analysis) There are several types of feature extraction techniques such as PLP,LPC and MFCC i noticed that the last one has performed a good results so Im gonna use this feature extraction technique: Mel-frequency Cepstral Coefficients (MFCC), this technique is one of the most important techniques used to extract the feature from an audio signal, it is actually based over the humans ear scale bandwidth. It uses the low and the high frequencies, measured in Hertz (Hz) to get the speech signal.  I did not chose just due to its popularity but also it is a perfect representation for sounds like music that is generated by humans and it has reduced vulnerability to noise disturbance.
    

```python
#CSV to save our extracted features
file_header = 'filename chroma_stft rmse spectral_centroid spectral_bandwidth rolloff zero_crossing_rate'
for i in range(1, 21):
    file_header += f' mfcc{i}'
file_header = file_header.split()
```

***Afeter diging into the second dataset for genre classification part i decide to pre-process the audio in a efficient way i will extract some other The spectral features which is obtained by converting the time-based signal into the frequency domain using the Fourier Transform, like :
- Spectral Centroid :the present the ferquency where the energy of a spectrum is centered or where the  center of mass for a sound is located.
- Zero Crossing Rate: A good way to measure the smoothness of a signal.
- Chroma Frequencies: It provides a robust way to describe a similarity measure between music pieces.
- Spectral Roll-off:It is a measure of the shape of the signal. It represents the frequency at which high frequencies decline to 0.
- spectral_bandwidth: is defined as the width of the band of light at one-half the peak maximum 

```python
file = open('dataset_e.csv', 'w', newline='')
with file:
    file_writer = csv.writer(file)
    file_writer.writerow(file_header)
for filename in os.listdir(f'part_2_raw_audio/'):
    songname = f'part_2_raw_audio/{filename}'
    y, sr = librosa.load(songname, mono=True, duration=30)
    rmse = librosa.feature.rms(y=y)
    chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
    spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
    spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    zcr = librosa.feature.zero_crossing_rate(y)
    mfcc = librosa.feature.mfcc(y=y, sr=sr)
    to_append = f'{filename} {np.mean(chroma_stft)} {np.mean(rmse)} {np.mean(spec_cent)} {np.mean(spec_bw)} {np.mean(rolloff)} {np.mean(zcr)}'    
    for e in mfcc:
        to_append += f' {np.mean(e)}'
    to_append += f''
    file = open('dataset_e.csv', 'a', newline='')
    with file:
        file_header = csv.writer(file)
        file_header.writerow(to_append.split())
```

```python
#read CSV into dataframe with pandas
df = pd.read_csv ('part3_GrooverChallengeDataset.csv')
df
```

```python
# Separate Features from the Target(label)
label_y=df["label"]
df_X = df.drop(columns=["label"], axis=1)
#Dropping unneccesary columns
df_X = df_X.drop(columns=["length"], axis=1)
df_X
```

```python
#Encoding the Labels (numbers from 0 to 9 because we 10 music classes)
genre_n = {}
k = 0
for genre in label_y.unique():
    genre_n[genre] = k
    k += 1
label_y = LabelEncoder().fit_transform(label_y)
#label_y
```

```python
# Split into training(75%) and testing sets(25%)
X_train, X_test, y_train, y_test = train_test_split(df_X, label_y, random_state=0)
```

```python
# Instantiate the model with RandomizedSearchCV optimized hyperparameters
rf_model = RandomForestClassifier(n_estimators=800, min_samples_split=5, min_samples_leaf=1,
                                     max_features="sqrt", max_depth=90, bootstrap=False)
```

```python
# Fit the model
rf_model = rf_model.fit(X_train, y_train)
```

```python
# Make predicitions using the testing data
rf_pred = rf_model.predict(X_test)
results = pd.DataFrame({
    "Predicted": rf_pred,
    "Correct": y_test
}).reset_index(drop=True)
results
```
